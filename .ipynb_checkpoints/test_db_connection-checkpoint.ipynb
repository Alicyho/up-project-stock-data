# PostgreSQL / pgAdmin 4 連線測試

此 Notebook 由 `test_db_connection.py` 轉成。

只測試「Python 是否能連到 PostgreSQL」，
不會抓股票資料，也不會修改資料庫。

## 步驟 0：安裝套件（裝過可略過）

```python
# !pip install sqlalchemy psycopg2-binary
```

## 步驟 1：設定連線

這一行要和 pgAdmin 4 的資料庫設定一致：
`Server -> MyLocalDB -> Databases -> stock_db`

```python
from sqlalchemy import create_engine, text

DATABASE_URL = "postgresql+psycopg2://postgres:Admin12345@localhost:5432/stock_db"
```

## 步驟 2：測試連線

成功會看到「連線成功！」以及目前資料庫名稱與 PostgreSQL 版本。

```python
def test_connection():
    """測試是否可以成功連到 PostgreSQL。"""

    try:
        # create_engine 會建立資料庫連線設定。
        engine = create_engine(DATABASE_URL)

        # engine.connect() 會真的嘗試連到 PostgreSQL。
        with engine.connect() as connection:
            # 問 PostgreSQL：我現在連到哪一個資料庫？
            database_name = connection.execute(
                text("SELECT current_database();")
            ).scalar()

            # 取得 PostgreSQL 版本資訊。
            postgres_version = connection.execute(
                text("SELECT version();")
            ).scalar()

        print("連線成功！")
        print(f"目前資料庫：{database_name}")
        print(f"PostgreSQL 版本：{postgres_version}")

    except Exception as error:
        print("連線失敗。")
        print("請檢查以下幾件事：")
        print("1. PostgreSQL 是否已經啟動")
        print("2. pgAdmin 4 裡是否真的有 stock_db")
        print("3. 使用者名稱是否是 postgres")
        print("4. 密碼是否是 Admin12345")
        print("5. Port 是否是 5432")
        print()
        print(f"錯誤訊息：{error}")


test_connection()
```
